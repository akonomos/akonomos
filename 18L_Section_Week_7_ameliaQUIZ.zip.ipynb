```python
#generating sound example
import numpy as np
import sounddevice
T = 3 # total number of seconds
fs = 44100 #sample rate

freq1 = 440 # 440 Hz frequency
freq2 = 442 # 442 Hz frequency

t = np.linspace(0, T, T*fs) # define our time vector
# IMPORTANT! Make sure the amplitudes are small
# (V < 0.3) to avoid damage to your computer speakers/headphones!
# define a perfect sine wave at freq1:
sine_waveform = (0.1)*np.sin(2*np.pi*freq1*t)

# define the sum of a sine wave at freq1
# and a sine wave at freq2:
beat_waveform = (0.1)*( np.sin (2*np.pi*freq1*t)+ np.sin (2*np.pi*freq2*t) )

sounddevice.play(sine_waveform , fs, blocking=True)

sounddevice.play(beat_waveform , fs)
```

```python
#computing fourier transform example
%matplotlib inline
import matplotlib.pyplot as plt

fft = np.fft.fft(beat_waveform) # Get FFT
freq_bins = np.fft.fftfreq(len(beat_waveform), d=1./fs)
NumSamples = beat_waveform.size

ampl_spctrm = np.abs(fft) # Get amplitude spectrum

# These next lines put them into proper units
ampl_spctrm = (np.sqrt(T)/NumSamples)*ampl_spctrm
pwr_spctrm = ampl_spctrm **2

fig , ax = plt.subplots ()
ax.set_xlabel('Frequency (Hz)')

ax.set_ylabel(r'Amplitude Spectrum $(V/\sqrt{\ mathrm{Hz}})$')

ax.axis ([400, 500, -0.01, 0.1])
ax.plot(freq_bins , ampl_spctrm)
```

```python
#functions to plot the amplitude spectrum and spectrograms of audio recordings
#computing fourier transform example
#generating sound example

T = 3 # total number of seconds
fs = 44100 #sample rate
freq1 = 440 # 440 Hz frequency
freq2 = 442 # 442 Hz frequency
t = np.linspace(0, T, T*fs) # define our time vector
# IMPORTANT! Make sure the amplitudes are small
# (V < 0.3) to avoid damage to your computer speakers/headphones!
# define a perfect sine wave at freq1:
sine_waveform = (0.1)*np.sin(2*np.pi*freq1*t)
beats = (0.1)*( np.sin (2*np.pi*freq1*t)+ np.sin (2*np.pi*freq2*t) )


%matplotlib inline
import matplotlib.pyplot as plt


fft = np.fft.fft(beats) # Get FFT
freq_bins = np.fft.fftfreq(len(beats), d=1./fs)
NumSamples = beats.size
ampl_spctrm = np.abs(fft) # Get amplitude spectrum
    
ampl_spctrm = (np.sqrt(T)/NumSamples)*ampl_spctrm
pwr_spctrm = ampl_spctrm **2
    
fig , ax = plt.subplots ()
ax.set_xlabel('Frequency (Hz)')
    
ax.set_ylabel(r'Amplitude Spectrum $(V/\sqrt{\ mathrm{Hz}})$')

#gives signal as a spectrogram
ax.axis ([400, 500, -0.01, 0.1])
ax.plot(freq_bins , ampl_spctrm)

 
```

```python

#import the pyplot and wavfile modules 
import matplotlib.pyplot as plot
from scipy.io import wavfile

samplingFrequency, signalData = wavfile.read('Physics_18L_Sample_Audio_File1.wav')

 
plot.subplot(211)
plot.title('Spectrogram of a wav file from canvas')

plot.plot(signalData)
plot.xlabel('Sample')
plot.ylabel('Amplitude')


plot.subplot(212)
plot.specgram(signalData,Fs=samplingFrequency)
plot.xlabel('Time')
plot.ylabel('Frequency')
plot.show()
```

```python
#function to reverse audio 
def ReverseAudio(Filename):
    fs, sound = read(Filename)
    reverse_sound = np.flip(sound, 0)
    new_name = Filename+ "Reverse.wav"
    write(new_name, fs, reverse_sound)
    sounddevice.play(reverse_sound, fs, blocking=True)
    return

ReverseAudio('Physics_18L_Sample_Audio_File1.wav')
```

```python
#speed up audio
def SpeedUp(Filename, speed):
    fs, sound = read(Filename)
    new_fs = speed * fs
    new_name = Filename+ "Speed.wav"
    write(new_name, new_fs, sound)
    sounddevice.play(sound, new_fs, blocking=True)
    return

SpeedUp('Physics_18L_Sample_Audio_File1.wav', 2)
```

```python
#slow down audio
def SlowDown(Filename, slowing):
    fs, sound = read(Filename)
    new_fs = int(1/slowing * fs)
    new_name = Filename+ "slow.wav"
    write(new_name, new_fs, sound)
    sounddevice.play(sound, new_fs, blocking=True)
    
SlowDown('Physics_18L_Sample_Audio_File1.wav', 2)
```
